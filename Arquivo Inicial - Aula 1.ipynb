### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
import pyautogui
import time

pyautogui.PAUSE=0.5

pyautogui.press('win')
pyautogui.write('chrome')
pyautogui.press('Enter')

time.sleep(2)

link='https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga'
pyautogui.write(link)
pyautogui.press('Enter')
time.sleep(3)
```

### Vamos agora enviar um e-mail pelo gmail

```python
import pyautogui
import time
import pyperclip

#abrir uma aba nova
pyautogui.hotkey('ctrl','t')
time.sleep(2)

#entrar no email
pyautogui.write('http://mail.google.com/')
pyautogui.press('Enter')
time.sleep(8)
pyautogui.click(x=74, y=189) #posicao do botao escrever no email
time.sleep(6)
pyautogui.write('gu.ger88+codigo@gmail.com')

#enviar o email
#  
pyautogui.press('tab') #seleciona o email
pyautogui.press('tab') #vai p o campo assunto

#campo assunto
pyperclip.copy('Relatório de Vendas')
pyautogui.hotkey('ctrl', 'v')
pyautogui.press('tab')

# corpo do email

texto = f'''
Prezados,

Segue o relatório de vendas de hoje,

O faturamento foi de = {faturamento}
A quantidade de produtos vendidos foi de = {quantidadeProdutos}

Qualquer dúvida estou a disposição.

Gustavo Germano'''

pyperclip.copy(texto)
pyautogui.hotkey('ctrl', 'v')

pyautogui.click(x=1307, y=999)#clicar no botao de envio de email
```
